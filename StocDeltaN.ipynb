```
As = 2.189 10^-9;
\[CapitalPi]2 = 50;
M = 0.1;
\[Phi]c = 0.1 Sqrt[2];
\[Mu]2 = 11;
\[Mu]1 = \[CapitalPi]2/(M^2 \[Phi]c)
\[CapitalLambda]4 = 12 \[Pi]^2 As/\[Mu]1^2
```

```
\[Psi]0 = 
 Sqrt[(Sqrt[2] \[CapitalLambda]4 M \[Phi]c^(1/2) \[Mu]1^(1/2))/(
  96 \[Pi]^(3/2))]
\[Chi]2 = Log[(\[Phi]c^(1/2) M)/(2 \[Mu]1^(1/2) \[Psi]0)]
```

```
N1 = (Sqrt[\[Chi]2] M \[Phi]c^(1/2) \[Mu]1^(1/2))/2
N2 = (M \[Phi]c^(1/2) \[Mu]1^(1/2))/(4 Sqrt[\[Chi]2])
```

```
\[Xi]k[NN_] = -((N1 + N2 - NN)/(\[Phi]c \[Mu]1));
\[Chi]k[NN_] = (4 \[Phi]c \[Mu]1 \[Xi]k[NN]^2)/M^2;
\[Psi]k[NN_] = \[Psi]0 Exp[\[Chi]k[NN]];
```

```
calPCG[NN_] = (\[CapitalLambda]4 M^2 \[Phi]c \[Mu]1)/(
  192 \[Pi]^2 \[Chi]2 \[Psi]k[NN]^2);
```

```
intcalP = Interpolation[meancalP];
```

```
n = 2;
```

```
field = {rho, theta};
fielddot = {rhodot, thetadot};
```

```
mm = 10^-5;
```

```
V[field[[1]]_, field[[2]]_] = 1/2 mm^2(field[[2]]^2 - 2/(3 field[[1]]^2)) 
metric[field[[1]]_, field[[2]]_] = DiagonalMatrix[{1,field[[1]]^2}]
```

```
inversemetric[field[[1]]_, field[[2]]_] = Inverse[metric[field[[1]],field[[2]]]]
```

```
affine[field[[1]]_,field[[2]]_] = Table[Sum[1/2 inversemetric[field[[1]],field[[2]]][[i,l]] (D[metric[field[[1]],field[[2]]][[j,l]],field[[k]]] + D[metric[field[[1]],field[[2]]][[k,l]],field[[j]]] - D[metric[field[[1]],field[[2]]][[j,k]],field[[l]]]), {l,n}], {i,n},{j,n},{k,n}]
```

```
H[field[[1]]_,field[[2]]_,fielddot[[1]]_,fielddot[[2]]_] = Sqrt[(V[field[[1]],field[[2]]] + 1/2 Sum[metric[field[[1]],field[[2]]][[i,j]] fielddot[[i]] fielddot[[j]], {i,n},{j,n}])] /. {mm -> 10^-5}
```

```
rhoi = 1.2356;
thetai = 13.5908;
tend = 10^5 mm^-1;
```

```
sol = NDSolve[{rho''[t] + affine[rho[t],theta[t]][[1,1,1]] rho'[t]^2 + 2 affine[rho[t],theta[t]][[1,1,2]] rho'[t] theta'[t] + affine[rho[t],theta[t]][[1,2,2]] theta'[t]^2 + 3 H[rho[t],theta[t],rho'[t],theta'[t]] rho'[t] + inversemetric[rho[t],theta[t]][[1,1]] D[V[rho[t],theta[t]], rho[t]] + inversemetric[rho[t],theta[t]][[1,2]] D[V[rho[t],theta[t]], theta[t]] == 0,
        theta''[t] + affine[rho[t],theta[t]][[2,1,1]] rho'[t]^2 + 2 affine[rho[t],theta[t]][[2,1,2]] rho'[t] theta'[t] + affine[rho[t],theta[t]][[2,2,2]] theta'[t]^2 + 3 H[rho[t],theta[t],rho'[t],theta'[t]] theta'[t] + inversemetric[rho[t],theta[t]][[2,1]] D[V[rho[t],theta[t]], rho[t]] + inversemetric[rho[t],theta[t]][[2,2]] D[V[rho[t],theta[t]], theta[t]] == 0,
        rho[0] == rhoi, theta[0] == thetai, rho'[0] == 0, theta'[0] == Sqrt[2/3] mm/rhoi/rhoi,
              NN'[t] == H[rho[t],theta[t],rho'[t],theta'[t]], NN[0] == 0}, {rho[t],theta[t],NN[t]}, {t,0,tend}] [[1]]
```
